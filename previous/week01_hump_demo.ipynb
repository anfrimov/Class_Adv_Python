```python
def hump(y0, b, n):
    """
    Generate a hump time series of length n with starting value t0 and tunung paramter b.
    """
    y = np.full(n, 0.)
    y[0] = y0
    for t in range(1, n):
        y[t] = b * y[t-1] * (1 - y[t-1])
    
    return y
```

Test it out

```python
y = hump(0.5, 3, 30)
print(y)
plt.plot(y)
```
